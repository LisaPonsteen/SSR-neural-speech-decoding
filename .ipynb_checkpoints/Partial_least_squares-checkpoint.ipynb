```python
# 1. Add this magic command at the very top of the cell
%matplotlib inline

import os
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, clear_output # For real-time updates
from scipy.stats import mannwhitneyu

# Helper function to calculate Cohen's d effect size
def cohen_d(x, y):
    nx, ny = len(x), len(y)
    if nx == 0 or ny == 0:
        return 0
    dof = nx + ny - 2
    pooled_std = np.sqrt(((nx - 1) * np.var(x, ddof=1) + (ny - 1) * np.var(y, ddof=1)) / dof)
    if pooled_std == 0:
        return 0
    return (np.mean(x) - np.mean(y)) / pooled_std

    
pts = ['sub-01', 'sub-02', 'sub-03', 'sub-04', 'sub-05', 'sub-06', 'sub-07', 'sub-08', 'sub-09', 'sub-10']

fig, axes = plt.subplots(2, 5, figsize=(20, 8), sharex=True, sharey=True)
axes = axes.flatten()

for i, pt in enumerate(pts):
    ax = axes[i]
    
    r, weights = reconstruct([pt], model='PLS', n_comp=9, feat_suffix='_onlyAMP_SSPEfeat.npy', unstacked=True, saveAs='dont')
    initParams_list = get_initParams(pt)
    all_osc_freqs = []
    for ch_params in initParams_list:
        all_osc_freqs.extend(ch_params['freqs'])
    
    all_osc_freqs = np.array(all_osc_freqs)
    all_stacked_freqs = np.tile(all_osc_freqs, 9)
    importance = np.mean(np.abs(weights), axis=0)

    '''hg_mask = (all_stacked_freqs >= 70) & (all_stacked_freqs <= 170)
    hg_vals = importance[hg_mask]
    non_hg_vals = importance[~hg_mask]
    
    mean_hg = np.mean(hg_vals) if len(hg_vals) > 0 else 0
    mean_non_hg = np.mean(non_hg_vals) if len(non_hg_vals) > 0 else 0

    # NEW: Calculate the 95th percentile to track the high outliers
    p95_hg = np.percentile(hg_vals, 95) if len(hg_vals) > 0 else 0
    p95_non_hg = np.percentile(non_hg_vals, 95) if len(non_hg_vals) > 0 else 0
    
    # --- Plotting the High-Outlier Indicators ---
    # Draw a solid, thin line showing where the top 5% of features start in each band
    if len(hg_vals) > 0:
        ax.hlines(p95_hg, 70, 170, colors='#e53e3e', linestyles='-', linewidth=1.2, zorder=3, label='Top 5% HG' if i==0 else "")
    ax.hlines(p95_non_hg, 0, 70, colors='#4a5568', linestyles='-', linewidth=1.2, zorder=3)
    ax.hlines(p95_non_hg, 170, 180, colors='#4a5568', linestyles='-', linewidth=1.2, zorder=3, label='Top 5% Non-HG' if i==0 else "")
    '''
    #print(f"[{pt}] Original Osc Count: {len(all_osc_freqs)} | Model Feature Count: {len(importance)}")

    # Draw background HG area shading using an unobtrusive neutral gray tint
    ax.axvspan(30, 70, color='C7', alpha=0.05, zorder=1)
    ax.axvspan(70, 170, color='C0', alpha=0.05, zorder=1)
    # Plot onto the specific subplot axis
    ax.scatter(all_stacked_freqs, importance, alpha=0.4, s=5, color='C' + str(0)) #blue
    
    ax.set_title(pt, fontsize=12)
    ax.set_xlim([0, 180])
    ax.grid(True, which='both', linestyle='--', alpha=0.5)
    
    # --- JUPYTER TRICK: Update the plot in real-time as subjects finish ---
    clear_output(wait=True)
    display(fig)

# 3. Add global X and Y labels
fig.text(0.5, 0.02, 'SSR Oscillator Frequency (Hz)', ha='center', fontsize=14)
fig.text(0.01, 0.5, 'Mean Absolute PLS Weight', va='center', rotation='vertical', fontsize=14)

plt.tight_layout(rect=[0.02, 0.05, 1, 0.95])

# Save and do a final explicit show
plt.savefig(os.path.join(result_path, 'PLS_weights_all_subjects_shaded.png'), dpi=300)
plt.show()
```
