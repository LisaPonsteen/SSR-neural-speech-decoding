```python


def plot_participant_model_fit(pt, feat_path, path_bids):
    """
    Plots the Grand Average PSD across all channels against a Histogram 
    of all fitted oscillator central frequencies for a given participant.
    """
    # 1. Load the initial parameters to get the number of channels and oscillator frequencies
    initParams_list = get_initParams(pt) # Assuming this function is available in your workspace
    num_channels = len(initParams_list)
    
    # Pool all oscillator central frequencies across all channels
    all_fitted_freqs = []
    for ch_idx in range(num_channels):
        freqs = initParams_list[ch_idx]['freqs']
        all_fitted_freqs.extend(freqs)
    all_fitted_freqs = np.array(all_fitted_freqs)
    
    # 2. Load the raw continuous iEEG data from the NWB file to compute the actual PSD
    nwb_path = os.path.join(path_bids, pt, 'ieeg', f'{pt}_task-wordProduction_ieeg.nwb')
    with NWBHDF5IO(nwb_path, 'r') as io:
        nwbfile = io.read()
        raw_data = nwbfile.acquisition['iEEG'].data[:]
        
    print(f"Data shape: {raw_data.shape}, Sampling Rate: {sr} Hz")
    print(f"Total oscillators collected across {num_channels} channels: {len(all_fitted_freqs)}")

    # 3. Compute the PSD for every channel using Welch's method
    # Adjust nperseg depending on desired frequency resolution (e.g., sr * 2 gives 0.5 Hz resolution)
    nperseg = 2048#int(sr * 2) 
    
    # Run once on channel 0 to get the frequency axis mapping
    full_freq_axis, temp_psd = scipy.signal.welch(raw_data[:, 0], fs=sr, nperseg=nperseg)
        # Find the indices that correspond to your desired 0-200 Hz range
    freq_mask = (full_freq_axis >= 0) & (full_freq_axis <= 200)
    freq_axis = full_freq_axis[freq_mask]  # This is your new 0-200Hz x-axis

    # Initialize your matrix using the length of the new clipped frequency axis
    all_psds = np.zeros((len(freq_axis), num_channels))
    all_psds[:, 0] = temp_psd[freq_mask]  # Store the clipped channel 0 PSD
        
    # Compute for the remaining channels
    for ch_idx in range(1, num_channels):
        _, psd = scipy.signal.welch(raw_data[:, ch_idx], fs=sr, nperseg=nperseg)
        all_psds[:, ch_idx] = psd[freq_mask]
        
    # Calculate the grand average and standard error across all channels
    grand_avg_psd = np.mean(all_psds, axis=1)
    sem_psd =(all_psds).std(axis=1) / np.sqrt(num_channels)

    # 4. Generate the Dual-Axis Plot
    fig, ax1 = plt.subplots(figsize=(10, 6))

    # --- Left Y-Axis: Empirical PSD (Line Plot) ---
    color_psd = 'tab:blue'
    ax1.set_xlabel('Frequency (Hz)', fontsize=12)
    ax1.set_ylabel('Power Spectral Density ($\mu V^2/Hz$)', color=color_psd, fontsize=12)
    
    # Plot grand average line
    ax1.plot(freq_axis, grand_avg_psd, color=color_psd, linewidth=2, label='Grand Avg PSD')
    # Plot shaded standard error bounding
    ax1.fill_between(freq_axis, grand_avg_psd - sem_psd, grand_avg_psd + sem_psd, 
                     color=color_psd, alpha=0.2)
    ax1.tick_params(axis='y', labelcolor=color_psd)
    ax1.set_yscale('log') # Power spectra are usually best viewed on a log scale

    # --- Right Y-Axis: SSR Model Oscillator Frequencies (Histogram) ---
    ax2 = ax1.twinx()  # Instantiate a second axes that shares the same x-axis
    color_hist = 'tab:orange'
    ax2.set_ylabel('Oscillator Central Frequency Count', color=color_hist, fontsize=12)
    
    # Define your histogram bins based on your frequency range of interest
    # e.g., bins from 0 to 120 Hz with 1 Hz step size
    bins = np.arange(0, 200, 1)
    
    ax2.hist(all_fitted_freqs, bins=bins, color=color_hist, alpha=0.5, 
             edgecolor='darkorange', label='Fitted Oscillators')
    ax2.tick_params(axis='y', labelcolor=color_hist)

    # --- Formatting and Limits ---
    ax1.set_xlim(0, 110) # Clip x-axis to your max tracking frequency (e.g. 100-110 Hz)
    plt.title(f'SSR Model Fit Validation: Participant {pt} ({num_channels} Channels Combined)', fontsize=14, pad=15)
    
    # Combine legends from both axes
    lines1, labels1 = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper right')

    plt.tight_layout()
    plt.show()

# Example invocation (Replace paths with your real variables):
# feat_path = '/path/to/features'
# path_bids = '/Users/lisa/Documents/DSAI_year2/Marble/SingleWordProductionDutch/SingleWordProductionDutch-iBIDS'
plot_participant_model_fit('sub-01', feat_path, path_bids)
```

```python
# 1. Add this magic command at the very top of the cell
%matplotlib inline

import os
import numpy as np
import scipy.signal
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
from pynwb import NWBHDF5IO

pts = ['sub-01', 'sub-02', 'sub-03', 'sub-04', 'sub-05', 'sub-06', 'sub-07', 'sub-08', 'sub-09', 'sub-10']

# Dictionary to hold all r-values group-by-participant
pt_r_values = {pt: [] for pt in pts}

# Pre-defined parameters from your envelope extraction logic
winL = 0.05
frameshift = 0.01
winL = 0.05

sr=1024

# Small helper function to speed up the hilbert transform
hilbert3 = lambda x: scipy.signal.hilbert(x, scipy.fftpack.next_fast_len(len(x)), axis=0)[:len(x)]

# --- 1. Compute Correlation Coefficients across ALL Channels ---
for pt in pts:
    print(f"Processing correlation arrays for {pt}...")
    
    # Load feature arrays
    hga = np.load(os.path.join(feat_path, f'{pt}_unstacked_feat.npy'))
    sspe = np.load(os.path.join(feat_path, f'{pt}_onlyAMP_SSPEfeat.npy'))
    initParams_list = get_initParams(pt)
    num_channels = len(initParams_list)
    
    # Load raw data to compute the traditional bandpass reference envelopes
    nwb_path = os.path.join(path_bids, pt, 'ieeg', f'{pt}_task-wordProduction_ieeg.nwb')
    with NWBHDF5IO(nwb_path, 'r') as io:
        nwbfile = io.read()
        data = nwbfile.acquisition['iEEG'].data[:]
        
    # Reconstruct the traditional Hilbert envelope space
    data_amp = np.abs(hilbert3(data))
    numWindows = int(np.floor((data.shape[0] - winL * sr) / (frameshift * sr)))
    raw_envelopes = np.zeros((numWindows, data.shape[1]))
    
    for win in range(numWindows):
        start = int(np.floor((win * frameshift) * sr))
        stop = int(np.floor(start + winL * sr))
        raw_envelopes[win, :] = np.mean(data_amp[start:stop, :], axis=0)
        
    # Iterate through channels to isolate and correlate high-gamma components
    start_col = 0
    for ch_idx in range(num_channels):
        freqs = initParams_list[ch_idx]['freqs']
        current_n_osc = len(freqs)
        
        # Isolate the traditional bandpass HG envelope feature
        feat_hga = hga[:, ch_idx]
        
        # Isolate individual oscillator matrix columns for this channel
        individual_oscs = sspe[:, start_col : start_col + current_n_osc]
        start_col += current_n_osc
        
        freqs = np.array(initParams_list[ch_idx]['freqs'])
        
        # 1. Create a boolean mask for the High Gamma band
        hg_mask = (freqs >= 70) & (freqs <= 170)
        
        # 2. Extract the correct oscillator index safely
        if np.any(hg_mask):
            # If multiple oscillators are in range, pick the one closest to 90 Hz (or your target center)
            target_center = 90.0
            in_range_indices = np.where(hg_mask)[0]
            hg_osc_idx = in_range_indices[np.argmin(np.abs(freqs[in_range_indices] - target_center))]
        else:
            # Fallback fallback: if none match the mask, grab the absolute highest frequency
            hg_osc_idx = np.argmax(freqs)
            print(f"Warning: No explicit HG oscillator found in range for {pt} Ch {ch_idx}. Selected {freqs[hg_osc_idx]:.1f} Hz.")

        # 3. Slice the individual oscillator matrices using your safe index
        feat_sspe_hg = individual_oscs[:, hg_osc_idx]
        
        # Ensure lengths match across extracted features before correlating
        min_len = min(len(feat_hga), len(feat_sspe_hg))
        
        # Calculate Pearson r correlation coefficient
        if np.std(feat_hga[:min_len]) > 0 and np.std(feat_sspe_hg[:min_len]) > 0:
            r_matrix = np.corrcoef(feat_hga[:min_len], feat_sspe_hg[:min_len])
            r_val = r_matrix[0, 1]
            pt_r_values[pt].append(r_val)
        else:
            pt_r_values[pt].append(0.0) # Handle zero-variance dropouts

# --- 2. Generate the Multi-Color Stacked Histogram ---
fig, ax = plt.subplots(figsize=(10, 6))

# Set up fixed bins spanning from -1.0 to 1.0 (with 0.05 step sizing)
bins = np.arange(-1.0, 1.05, 0.05)

# Extract lists of data arrays and map names for the stacked layout
data_to_stack = [pt_r_values[pt] for pt in pts]
colors = plt.cm.tab10(np.linspace(0, 1, len(pts))) # Generates 10 distinct thematic colors

# Plot stacked histogram
ax.hist(data_to_stack, bins=bins, stacked=True, color=colors, alpha=0.8,
        edgecolor='white', linewidth=0.5, label=pts)

# Styling and Labels
ax.set_xlabel("Pearson Correlation Coefficient ($r$)", fontsize=13, fontweight='bold', labelpad=10)
ax.set_ylabel("Channel Counts", fontsize=13, fontweight='bold', labelpad=10)
ax.set_title("Feature Fidelity: Traditional Bandpass vs. SSR High-Gamma Envelopes", fontsize=14, fontweight='bold', pad=15)
ax.set_xlim([-0.2, 1.0]) # Clip x-axis around zero since features should correlate positively
ax.grid(True, which='major', linestyle='--', linewidth=0.5, alpha=0.3)

# Place legend cleanly on the side
ax.legend(title="Participants", bbox_to_anchor=(1.02, 1), loc='upper left', frameon=True, edgecolor='#cbd5e1')

plt.tight_layout()
plt.savefig(os.path.join(result_path, 'SSR_bandpass_correlation_stacked_hist.png'), dpi=300)
plt.show()
```

```python
duration_sec = 5
t_start = 10
feat_sr = 1 / 0.01  # if frameshift is 0.01s

# 2. Calculate indices based on the FEATURE clock
start_win = int(t_start * feat_sr)
# 2 seconds of data = 200 windows
num_wins = int(duration_sec * feat_sr) 
end_win = start_win + num_wins

# 3. Create a time vector that matches the length of the features (200,)
t = np.linspace(t_start, t_start + duration_sec, num_wins)

def get_feat_for_plt(pt, ch_idx):
    hga = np.load(os.path.join(feat_path,f'{pt}_unstacked_feat.npy'))
    sspe = np.load(os.path.join(feat_path,f'{pt}_onlyAMP_SSPEfeat.npy'))

    #print(f'hga.shape: {hga.shape}, sspe.shape: {sspe.shape}')
    
    
    feat_hga = hga[start_win : end_win, ch_idx].flatten()
    feat_hga_norm = (feat_hga - np.mean(feat_hga)) / np.std(feat_hga) # Normalize for visual comparison (z-score) so they sit on top of each other
    
    initParams_list = get_initParams(pt)
    initParams = initParams_list[ch_idx]
    freqs = initParams_list[ch_idx]['freqs']

    start_col = 0
    for i in range(ch_idx):
        start_col += len(initParams_list[i]['freqs'])
    
    # The number of oscillators for the current channel
    current_n_osc = len(freqs)

    # Get the individual oscillator columns for this channel
    individual_oscs = sspe[start_win : end_win, start_col : start_col + current_n_osc]
    
    # Sum them for the total SSPE feature
    feat_sspe = np.sum(individual_oscs, axis=1)
    
    if np.std(feat_sspe) ==0:
        print(f"std sspe: {np.std(feat_sspe)}, ch_idx: {ch_idx}")
        print('feat sspe ',feat_sspe)
        feat_sspe_norm = feat_sspe
    feat_sspe_norm = (feat_sspe - np.mean(feat_sspe)) / np.std(feat_sspe)

    
    # Normalize individual oscillators for comparison
    # return these as a list of normalized arrays
    indiv_oscs_norm = []
    for i in range(current_n_osc):
        osc = individual_oscs[:, i]
        if np.std(osc) > 0:
            indiv_oscs_norm.append((osc - np.mean(osc)) / np.std(osc))
        else:
            indiv_oscs_norm.append(osc) # handle zero-variance case
    
    #raw data features
    winL = 0.05
    frameshift = 0.01
    modelOrder = 4
    stepSize = 5
    path_bids = '/Users/lisa/Documents/DSAI_year2/Marble/SingleWordProductionDutch/SingleWordProductionDutch-iBIDS'
    io = NWBHDF5IO(os.path.join(path_bids,pt,'ieeg',f'{pt}_task-wordProduction_ieeg.nwb'), 'r')
    nwbfile = io.read()
    data = nwbfile.acquisition['iEEG'].data[:]
    io.close()
    
    #Small helper function to speed up the hilbert transform by extending the length of data to the next power of 2
    hilbert3 = lambda x: scipy.signal.hilbert(x, scipy.fftpack.next_fast_len(len(x)),axis=0)[:len(x)]
    
    numWindows = int(np.floor((data.shape[0]-winL*sr)/(frameshift*sr)))
 
    #Create feature space
    data = np.abs(hilbert3(data))
    feat = np.zeros((numWindows,data.shape[1]))
    for win in range(numWindows):
        start= int(np.floor((win*frameshift)*sr))
        stop = int(np.floor(start+winL*sr))
        feat[win,:] = np.mean(data[start:stop,:],axis=0)
    
    raw_envelope = feat[start_win : end_win, ch_idx]
    
    # Normalize for visual comparison so they sit on top of each other
    raw_envelope_norm = (raw_envelope - np.mean(raw_envelope)) / np.std(raw_envelope)

    
    return raw_envelope_norm, feat_hga_norm, feat_sspe_norm, np.array(indiv_oscs_norm), freqs
```
